```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# data should be in below format

# data dir
#     train dir
#         class 1 dir  # all dogs images
#         class 2 dir  # all cats images
#         class n dir
#     validation dir
#         class 1 dir
#         class 2 dir
#         class n dir
```

```python
root = os.path.join('data','img_data')  # data is main dir, img_data is sub-dir
train_data_path = os.path.join(root,'train')
valid_data_path = os.path.join(root,'validation')
```

```python
height,width,channel = (224,224,3)
IMG_SIZE = (height,width)

train_transform = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.ToTensor()
])  # compose applies list of transformations on the data

valid_transform = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.ToTensor()
])  # compose applies list of transformations on the data
```

```python
# first we'll get the data it will be from dataset module which
# has this ImageFolder method which gives option to get the data from
# image folder arranged like above mentioned
train_data = datasets.ImageFolder(root=train_data_path,
                                        transform=train_transform)
# then we'll load the data
train_data_loader = DataLoader(train_data, batch_size=10, shuffle=True)
```

```python
# train_data_loader will have iterative object,
# which will contain 2 items, 1. images, 2. labels
# images will contain no. batch size images of given resize
# labels will contain labels for all those images

data = next(iter(train_data_loader))
print(len(data))  # =2
print(data[0].shape)  # =(batch size, image_size)  actual images data
print(data[1].shape)  # =(bathc size)  labels for images
```
