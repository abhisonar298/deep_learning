```python
class Iris(Dataset):  # inheriting from parent Dataset class of torch
    def __init__(self, target_col='species'):  # set target column
        self.df = sb.load_dataset('iris')  # load the complete dataset
        
        x = self.df.drop(target_col, axis=1).to_numpy()  # set independant columns as x and in numpy array to convert it into tensor later
        self.x = torch.from_numpy(x)  # converting to tensor and setting as x
        
        replace_dict = {'setosa': 0, 'versicolor': 1, 'virginica': 2}  # target label encoding dictionary
        y = self.df[target_col].replace(replace_dict).to_numpy()  # label encoding of target column and in numpy array to convert it into tensor later
        self.y = torch.from_numpy(y)  # converting independant column into a tensor
        
    def __getitem__(self, index):  # creating method to fetch the independant and target column values for specified index
        return self.x[index], self.y[index]
    
    def __len__(self):  # method to return the total no. of rows of the dataset
        return self.df.shape[0]
```

```python
iris = Iris()
```

```python
len(iris)
```

```python
iris_data_loader = DataLoader(iris, batch_size=10, shuffle=True)
```

```python
for i in iris_data_loader:
    x,y = i
    print('x=',x)  # independant column
    print('y=',y)  # target column
    break
```
