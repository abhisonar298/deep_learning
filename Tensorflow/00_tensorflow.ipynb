### Shuffling

```python
a = tf.constant(
    [[1,2],
    [3,4],
    [4,5],
    [6,7],
    [8,9]]
)
```

```python
tf.random.set_seed(1)  # global level seed
tf.random.shuffle(a, seed=10)  # operation level seed

# when both global and operational seeds are fixed,
# the shuffling output  will be the same
```

```python
tf.random.set_seed(10)
tf.random.shuffle(a, seed=10)
```

```python
# tf.random.set_seed(10)
tf.random.shuffle(a, seed=10)
```

### Matrix Multiplication

```python
a = tf.constant(
    [[7,10,4],
    [10,7,6],
    [2,6,9]]
)
b = tf.constant(
    [[5,6],
    [8,2],
    [7,9]]
)
print(a)
print(b)
```

```python
tf.matmul(a,b)
```

```python
b, tf.transpose(b), tf.reshape(b, shape=[2,3])
```

```python
# transpose flips the axes, rows become columns and viseversa
# reshape re-orders elements
```

```python
tf.tensordot(a,b, axes=1)
```

```python
tf.matmul(a,b)
```

```python
a = tf.constant([
    [1,2,3],
    [4,5,6],
    [7,8,9]
])

b = tf.constant([
    [9,8,7],
    [6,5,4],
    [3,2,1]
])
```

```python
tf.matmul(tf.transpose(a),b)
```

```python
tf.matmul(a,tf.transpose(b))
```

### Aggregating Tensors

```python
# collective value
```

```python
a = tf.constant([
    [1,-2,3],
    [-4,5,-6],
    [7,-8,9]
])
a
```

```python
# min
# max
# mean
# sum
```

```python
tf.reduce_min(a)
```

```python
tf.reduce_max(a)
```

```python
tf.reduce_mean(a)
```

```python
tf.reduce_sum(a)
```

```python
# variance # Standard Deviation 
```
