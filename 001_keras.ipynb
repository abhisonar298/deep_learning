```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

# 1. Regression

```python
model = Sequential()
```

```python
import pandas as pd

dict1 = {
    'column1':[1,2,3,4,5],
    'column2':[6,7,8,9,10],
    'column3':[11,12,13,14,15]
}

df = pd.DataFrame(dict1)
df.shape[1]
```

Split the data X_train, y_train, X_test, y_test

```python
# model with 8 inputs, 2 hidden layers 5 nodes each and 1 output
model.add(keras.Input(shape=8))  # input layer
model.add(Dense(5, activation='relu'))  # hidden layer
model.add(Dense(5, activation='relu'))  # hidden layer
model.add(Dense(1))  # output layer
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

# 2. Classification

In multiclass classification need to convert the target data in to a binary array.

y_train = keras.utils.to_categorical(df['target_column'])

no. of output nodes = no. of target classes

```python
# model with 8 inputs, 2 hidden layers 5 nodes each and 1 output
model.add(keras.Input(shape=8))  # input layer
model.add(Dense(5, activation='relu'))  # hidden layer
model.add(Dense(5, activation='relu'))  # hidden layer
model.add(Dense(4, activation='softmax'))  # output layer
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=10)
predictions = model.predict(X_test)
```
